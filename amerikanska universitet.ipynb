**(f)** Create a new qualitative (yes/no) variable, called Elite, by binning the Top10perc variable into two groups based on whether or not the proportion of students coming from the top 10% of their high school classes exceeds 50%.

`college[â€™Eliteâ€™] = pd.cut(college[â€™Top10percâ€™], [0,0.5,1], labels
 =[â€™Noâ€™, â€™Yesâ€™])`

Use the `value_counts()` method of `college[â€™Eliteâ€™]`to see how many **elite** universities there are. Finally, use the `boxplot()` method again to produce side-by-side boxplots of Outstate versus Elite.

```python
#Create a new categorical column 'Elite' based on Top10perc (>50% = 'Yes', â‰¤50% = 'No')
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels=['No', 'Yes'])
display(college.head())
# Count how many colleges are 'Yes' (Elite) and 'No' (non-Elite)
college['Elite'].value_counts()
# Draw a boxplot comparing Out-of-state tuition by Elite status
sns.boxplot(x='Elite', y='Outstate', data=college)
plt.title('Out-of-State Tuition by Elite')
plt.xlabel('Elite')
plt.ylabel('Out-of-State Tuition')
```

The boxplot compares Out-of-state tuition between Elite and non-Elite colleges.
It shows that colleges classified as Elite **(Elite = Yes)** tend to have higher out-of-state tuition compared to non-Elite colleges **(Elite = No)**.<br>
This indicates that more selective or academically strong institutions usually charge higher tuition fees.<br>

**(g)** Use the `plot.hist()` method of college to produce some histograms with differing numbers of bins for a few of the quantitative variables. The command `plt.subplots(2, 2)` may be useful: it will divide the plot window into four regions so that four plots can be made simultaneously. By changing the arguments
you can divide the screen up in other combinations

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8)) # Create a 2*2 grid of subplots

# Plot histograms for four quantitative variables with different bin sizes
sns.histplot(college['Apps'], bins=15, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Applications (15 bins)')

sns.histplot(college['Enroll'], bins=25, kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Enrollment (25 bins)')

sns.histplot(college['Top10perc'], bins=10, kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Top 10% Students (10 bins)')

sns.histplot(college['Outstate'], bins=30, kde=True, ax=axes[1, 1], color='violet')
axes[1, 1].set_title('Out-of-State Tuition (30 bins)')
```

**The histograms display the distributions of several quantitative variables.**<br>
By changing the number of bins, we can observe different levels of details in the data. 
For example, `Outstate` tuition shows a right-skewed distribution, indicating that most colleges charge lower tuition, while a few have very high values. <br>
Similarly, `Apps` and `Enroll` are also right-skewed, meaning a small number of colleges receive or enroll far more students than most others.
